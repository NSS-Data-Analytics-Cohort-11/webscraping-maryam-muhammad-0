```python
URL_jd = 'https://realpython.github.io/fake-jobs/jobs/senior-python-developer-0.html'

response_jd = requests.get(URL_jd)
```

```python
webscr_jd = BS(response_jd.text)
```

```python
webscr_jd.findAll('p')[1].text
```

```python
def get_desc(url_jd):
    response_jd = requests.get(url_jd)
    webscr_jd = BS(response_jd.text)
    return webscr_jd.findAll('p')[1].text
```

```python
get_desc('https://realpython.github.io/fake-jobs/jobs/television-floor-manager-8.html')
```

```python
job_info['Job Description'] = job_info['URL'].apply(get_desc)
```
